```python
countries = ["India","Tokelau","Brazil"]
df.groupby('Name')['Cases - newly reported in last 7 days'].sum().loc[countries]
```

```python
regions = ['Americas', 'Europe','South-East Asia']
df.groupby('WHO Region')['Cases - newly reported in last 7 days'].agg(['mean','median','sum']).loc[regions]
```

```python
countries = ["India","Tokelau","Brazil"]
df.groupby('Name')['Cases - newly reported in last 7 days'].agg(['mean','median','sum']).loc[countries]
```
