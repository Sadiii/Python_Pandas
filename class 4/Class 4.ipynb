### apply()
- series and dataframe 

### DataFrame

```python
df = df.apply(lambda x: x.str.upper())
```

```python
df
```

```python
df = df.apply(lambda x: x.str.lower())
```

```python
df
```

### Series

```python
df['Country Code'] = df['Country Code'].apply(lambda x: x.upper())
```

```python
df
```

```python
df.apply(len)
```

```python
df['Country Name'].apply(len)
```

```python
df.apply(len, axis = 'rows')
```

```python
df.apply(len, axis = 'columns')
```

### Applymap()
- works only on dataframe

```python
df = df.apply(lambda x: x.str.upper())
```

```python
df.applymap(str.lower)
```

```python
df.applymap(len)
```

### map()
- series

```python
df['Country Name'] = df['Country Name'].map(str.lower)
```

```python
df
```
