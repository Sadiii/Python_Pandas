- slicing

### loc and iloc functions

```python
import pandas as pd
```

```python
df = pd.read_csv('data/covidcases.csv')
df
```

#### loc

#### list

```python
df.loc[[2,4,8],['Name', 'WHO Region', 'Cases - cumulative total']]
```

#### slicing

```python
df.loc[2:10,['Name', 'WHO Region', 'Cases - cumulative total']]
```

```python
df.loc[2:10,'Name':'Cases - cumulative total']
```

```python
df.loc[[2],['Name']]
```

### filtering 

```python
filt = df['WHO Region'] == 'Americas'
df.loc[filt, :]
```

```python
filt = df['WHO Region'] == 'Americas'
df.loc[filt, 'Name':'Cases - cumulative total']
```

### iloc

```python
df.iloc[1:10,1:5]
```

```python
df.iloc[[1,10],0:5]
```

```python
df.loc[df['WHO Region'] == 'Americas' , 'Name':'Cases - cumulative total']
```
